## 1. pylab 서브패키지

## 1.1. 라인 플롯
선을 그림
* 데이터가 시간, 순서 등에 따른 변화를 보여주는 용도

##### plot
라인 플롯을 그린다.
* 입력 : 리스트 혹은 `ndarray` 객체

```python
plt.title("Plot")
plt.plot([1, 4, 9, 16])
plt.yticks([1, 4, 9, 16])
plt.grid()
plt.show()
```

이 때 x축의 tick(x축 자료위치)은 자동으로 0, 1, 2, 3이 된다.
*  x tick 명시할 때는 2개의 같은 길이 리스트나 배열을 넣는다.

##### show

* 주피터 노트북에서는 셀 단위로 플롯 명령을 자동 렌더링 해주어 `show` 명령이 필요없음
* 일반 파이썬 인터프리터 환경을 대비해 항상 마지막에 실행되도록 한다.
* 마지막 플롯 명령에서 반환되는 플롯 객체의 표현도 가려주는 효과가 있다.

## 1.4. 틱 설정
틱(tick) : 플롯이나 차트에서 축상의 위치 표시 지점  
틱 라벨(tick label) : 틱에 써진 숫자 혹은 글자
##### xticks, yticks
틱 위치나 틱 라벨을 수동 설정한다.

```python
import numpy as np

X = np.linspace(-np.pi, np.pi, 256)
C = np.cos(X)

plt.title("x축과 y축의 tick label 설정")
plt.plot(X, C)
plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi])
plt.yticks([-1, 0, +1])
plt.grid()
plt.show()
```

틱 라벨 문자열에 `$$` 사이에 LaTeX 수학 문자식도 넣을 수 있다.

```python
X = np.linspace(-np.pi, np.pi, 256)
C = np.cos(X)

plt.title("LaTeX, 문자열로 tick label 설정")
plt.plot(X, C)
plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi],
           [r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi/2$', r'$+\pi$'])
plt.yticks([-1, 0, +1], ["Low", "Zero", "High"])
plt.grid()
plt.show()
```

## 1.6. 여러 개 선 그리기
라인플롯에서 여러 선을 그릴 때는 인수로 x 데이터, y 데이터, 스타일 문자열을 반복하여 넣는다.

## 1.7. 홀드 명령
##### hold(True|False)
복수의 `plot`을 한 그림에 겹쳐 그리기
* Matplotlib 1.5까지는 `hold(True)`로 기존 그림위에 겹치기 on, `hold(False)`로 겹치기 종료
* Matplotlib 2.0부터 모든 플롯에 `hold(True)` 자동 적용

```python
plt.title("복수의 plot 명령을 한 그림에서 표현")
plt.plot([1, 4, 9, 16],
        c="b", lw=5, ls="--", marker="o", ms=15, mec="g", mew=5, mfc="r")
# plt.hold(True) # 1.5 버전에서 필요한 코드
plt.plot([9, 16, 4, 1],
        c="k", lw=3, ls=":", marker="s", ms=10, mec="m", mew=5, mfc="c")
# plt.hold(False) # 1.5 버전에서 필요한 코드
plt.show()
```

## 1.8. 범례
##### legend
범례(legend)를 추가
* 범례 위치 수동 설정 시 `loc` 인수를 사용
* `loc` 인수에는 문자열 또는 숫자가 가능

| loc 문자열 | 숫자 |
|------------|------|
|`best` | 0 | 
|`upper right` | 1 | 
|`upper left` | 2 | 
|`lower left` | 3 |

```python
X = np.linspace(-np.pi, np.pi, 256)
C, S = np.cos(X), np.sin(X)
plt.title("legend를 표시한 플롯")
plt.plot(X, C, ls="--", label="cosine")
plt.plot(X, S, ls=":", label="sine")
plt.legend(loc=2)
plt.show()
```

### 1.10.1. Figure 객체
* `Matplotlib.figure.Figure` 클래스 객체에 포함됨
* 내부 플롯(inline plot)이 아닌 경우에는 하나의 Figure는 하나의 ID 숫자와 윈도우(Window)를 가진다.
* 주피터 노트북에서는 윈도우 객체가 생성되지 않는다.
* 파이썬을 독립 실행하는 경우 하나의 Figure당 하나의 윈도우를 별도로 가진다.
* http://Matplotlib.org/api/figure_api.html#Matplotlib.figure.Figure

##### figure
Figure를 생성한다.
* `plot` 명령 실행 시 자동으로 Figure를 생성해주기 때문에 `figure` 명령을 명시하는 경우는 많지 않다.
* `figure` 명령을 실행하는 경우
    * 여러 개의 윈도우를 동시에 띄울 때(line plot이 아닌 경우)
    * 주피터 노트북 등에서(line plot의 경우) 그림의 크기를 설정하고 싶을 때
* 그림의 크기를 `figsize` 인수로 설정

```python
np.random.seed(0)
f1 = plt.figure(figsize=(10, 2))
plt.title("figure size : (10, 2)")
plt.plot(np.random.randn(100))
plt.show()
```

##### gcf
현재 사용하고 있는 Figure 객체을 얻는다.
* 다른 변수에 할당도 가능

```python
f1 = plt.figure(1)
plt.title("현재 Figure 객체")
plt.plot([1, 2, 3, 4], 'ro:')

f2 = plt.gcf()
print(f1, id(f1))
print(f2, id(f2))
plt.grid()
plt.show()
```

### 1.10.2. Axes 객체와 `subplot` 명령
* http://Matplotlib.org/api/axes_api.html#Matplotlib.axes.Axes

##### subplot
명시적으로 그리드(grid) 형태의 Axes 객체를 생성
* 그러나 `plot` 명령으로 Axes 객체를 자동 생성해준다.
* Figure는 행렬(matrix), Axes는 행렬의 원소라고 생각하면 된다.  
  ex)  
  위아래 2개 플롯이 있는 경우 2 × 1 행렬
* `subplot(첫 번째 인수, 두 번째 인수, 세 번째 인수)`
    * 처음 2개 인수 : 전체 그리드 행렬의 모양 지시
        * 행, 열을 나타내는 숫자 기입
    * 세 번째 인수 : 플롯 수 중 어느 것인지를 의미하는 숫자
* 인수는 줄여 하나의 숫자로 표시 가능하다.  
  ex)  
  (2, 2, 1) → 221
* Axes의 위치는 위 → 아래, 왼쪽 → 오른쪽으로 카운트

##### tight_layout
플롯 간의 간격을 자동으로 맞춘다.

```python
x1 = np.linspace(0.0, 5.0)
x2 = np.linspace(0.0, 2.0)
y1 = np.cos(2 * np.pi * x1) * np.exp(-x1)
y2 = np.cos(2 * np.pi * x2)

ax1 = plt.subplot(2, 1, 1)
plt.plot(x1, y1, 'yo-')
plt.title("A tale of 2 subplots")
plt.ylabel("Damped oscillation")
print(ax1)

ax2 = plt.subplot(2, 1, 2)
plt.plot(x2, y2, 'r.-')
plt.xlabel("time (s)")
plt.ylabel("Undamped")
print(ax2)

plt.tight_layout()
plt.show()
```

```python
# 2 * 2의 4개 플롯 그리기
np.random.seed(0)

plt.subplot(221)
plt.plot(np.random.rand(5))
plt.title("axes 1")

plt.subplot(222)
plt.plot(np.random.rand(5))
plt.title("axes 2")

plt.subplot(223)
plt.plot(np.random.rand(5))
plt.title("axes 3")

plt.subplot(224)
plt.plot(np.random.rand(5))
plt.title("axes 4")

plt.tight_layout()
plt.show()
```

##### subplots
복수의 Axes 객체 동시 생성
* 이 때는 2차원 ndarray 형태로 Axes 객체가 반환됨

```python
fig, axes = plt.subplots(2, 2)

np.random.seed(0)

axes[0, 0].plot(np.random.rand(5))
axes[0, 0].set_title("axes 1")

axes[0, 1].plot(np.random.rand(5))
axes[0, 1].set_title("axes 2")

axes[1, 0].plot(np.random.rand(5))
axes[1, 1].set_title("axes 3")

axes[1, 1].plot(np.random.rand(5))
axes[1, 1].set_title("axes 4")

plt.tight_layout()
plt.show()
```

### 1.10.3. Axis 객체와 축
하나의 Axes 객체는 2개 이상의 Axis 객체를 가진다.  
Axis 객체는 플롯의 가로축이나 세로축을 나타내는 객체
* https://matplotlib.org/api/axis_api.html

##### twinx
복수의 y축을 가진 플롯을 만든다.
* `twinx`명령은 x축을 공유하는 새로운 Axes 객체를 만든다.

```python
fig, ax0 = plt.subplots()
ax1 = ax0.twinx()

ax0.set_title("2개의 y축 한 figure에서 사용하기")
ax0.plot([10, 5, 2, 9, 7], 'r-', label="y0")
ax0.set_ylabel("y0")
ax0.grid(False)
ax0.set_xlabel("공유되는 x축")

ax1.plot([100, 200, 220, 180, 120], 'g', label="y1")
ax1.set_ylabel("y1")
ax1.grid(False)

plt.show()
```

```python
X = np.linspace(-np.pi, np.pi, 256)
C, S = np.cos(X), np.sin(X)

plt.title("한글 제목")
plt.plot(X, C, label="코사인")
t = 2 * np.pi / 2

plt.scatter(t, np.cos(t), 50, color='blue')
plt.xlabel("엑스축 라벨")
plt.ylabel("와이축 라벨")
plt.annotate("여기가 0.5!", xy=(t, np.cos(t)), xycoords='data', xytext=(-90, -59),
            textcoords='offset points', fontsize=16, arrowprops=dict(arrowstyle="->"))
plt.grid()
plt.show()
```

### 2.1.2. 개별적으로 폰트 적용
폰트 패밀리, 색상, 크기를 정하여 플롯 명령들의 `fontdict` 인수에 넣는다.

```python
font1 = {'family': 'NanumMyeongjo', 'color': 'black', 'size': 24}
font2 = {'family': 'NanumBarunpen', 'color': 'darkred', 'weight': 'bold', 'size': 18}
font3 = {'family': 'NanumBarunGothic', 'color': 'blue', 'weight': 'light', 'size': 12}

x = np.linspace(0.0, 5.0, 100)
y = np.cos(2 * np.pi * x) * np.exp(-x)

plt.plot(x, y, 'k')
plt.title('한글 제목', fontdict=font1)
plt.xlabel('엑스 축', fontdict=font2)
plt.ylabel('와이 축', fontdict=font3)
plt.subplots_adjust()
plt.show()
```

# 1. 바 차트
##### bar
* x 데이터가 카테고리 값인 경우 세로 방향 바 차트(bar chart)를 그린다.
* 참고
    * http://matplotlib.org/api/pyplot_api.html#matplotlib.pyplot.bar
* `bar(left, height)`
    * `left` : x축에서 바(bar)의 왼쪽 변의 위치를 나타낸다.

```python
import matplotlib as mpl
import matplotlib.pylab as plt
import numpy as np

y = [2, 3, 1]
x = np.arange(len(y))
xlabel = ['가', '나', '다']

plt.title("Bar Chart")
plt.bar(x, y)
plt.xticks(x, xlabel)
plt.yticks(sorted(y))
plt.xlabel("가나다")
plt.ylabel("빈도 수")
plt.show()
```

#### barh
* x 데이터가 카테고리 값인 경우 가로 방향 바 차트(bar chart)를 그린다.
* 참고
    * http://matplotlib.org/api/pyplot_api.html#matplotlib.pyplot.barh
* `barh(bottom, width)`
    * `bottom` : y축
    * `xerr`, `yerr` : 에러 바(error bar) 추가
    * `alpha` : 투명도 지정
        * 0 : 완전 투명
        * 1 : 완전 불투명

```python
np.random.seed(0)

people = ["몽룡", "춘향", "방자", "향단"]
y_pos = np.arange(len(people))
performance = 3 + 10 * np.random.rand(len(people))
error = np.random.rand(len(people))

plt.title("Barh Chart")
plt.barh(y_pos, performance, xerr=error, alpha=0.4)
plt.yticks(y_pos, people)
plt.xlabel('x 라벨')
plt.grid()
plt.show()
```

# 2. 스템 플롯
* 스템 플롯(stem plot) : 폭이 없는 바 차트
* 주로 이산 확률 함수, 자기상관관계(auto-correlation) 묘사 시 사용
* 참고
    * http://matplotlib.org/api/pyplot_api.html#matplotlib.pyplot.stem
    
##### stem
스템 플롯을 그린다.

```python
x = np.linspace(0.1, 2 * np.pi, 10)
plt.title("Stem Plot")
plt.stem(x, np.cos(x), '-.')
plt.grid()
plt.show()
```

# 3. 파이 차트
* 파이 차트(pie chart)를 그릴 때는 원의 형태를 유지할 수 있도록 다음 명령을 실행해야 한다.
`plt.axis('equal')`
* 참고
    * http://matplotlib.org/api/pyplot_api.html#matplotlib.pyplot.pie

##### pie
파이 차트를 그린다.
* `explode` : None이 아니면 각 쐐기의 간격을 띄우기할 반지름의 비율을 지정하는 배열을 지정한다?
* `autopct` : 각 쐐기의 라벨값 지정하는 string이나 함수
* `startangle` : x축의 반시계 방향으로 설정 각도만큼 회전

```python
labels = ['개구리', '돼지', '개', '통나무']
sizes = [15, 30, 45, 10]
colors = ['yellowgreen', 'gold', 'lightskyblue', 'lightcoral']
explode = (0, 0.1, 0, 0)

plt.title("Pie Chart")
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%',
       shadow=True, startangle=90)
plt.axis('equal')
plt.show()
```

# 4. 히스토그램
* 참고
    * http://matplotlib.org/api/pyplot_api.html#matplotlib.pyplot.hist

##### hist
히스토그램을 그린다.
* `hist(x)`
    * `bins` : 데이터를 집계할 구간 정보 설정
        * 반환값 : 데이터 집계 결과
* 반환값  
(n, bins, patches) 튜플
    * `n` : 히스토그램 bins의 값, 빈도수
    * `bins` : bins의 가장자리
    * `patches` : 개별 패치의 목록

```python
np.random.seed(0)
x = np.random.randn(1000)

plt.title("Histogram")
arrays, bins, patches = plt.hist(x, bins=10)
plt.grid()
plt.show()
```

# 5. 스캐터 플롯
* 참고
    * http://matplotlib.org/api/pyplot_api.html#matplotlib.pyplot.scatter
    
##### scatter
스캐터 플롯을 그린다.

```python
np.random.seed(0)
X = np.random.normal(0, 1, 100)
Y = np.random.normal(0, 1, 100)

plt.title("Scatter Plot")
plt.scatter(X, Y)
plt.show()
```

데이터가 3차원, 4차원인 경우에는 점의 크기나 색깔을 이용하여 나타낼 수 있다.  
이런 차트를 **버블 차트(bubble chart)**라고 한다.
* 인수
    * `s` : 크기
    * `c` : 색깔

```python
N = 30
np.random.seed(0)
x = np.random.rand(N)
y1 = np.random.rand(N)
y2 = np.random.rand(N)
y3 = np.pi * (15 * np.random.rand(N)) ** 2

plt.title("Bubble Chart")
plt.scatter(x, y1, c=y2, s=y3)
plt.show()
```

# 6. Imshow
* 참고
    * http://matplotlib.org/api/pyplot_api.html#matplotlib.pyplot.imshow
    
##### imshow
화상(image) 데이터처럼 행, 열을 가진 행렬 형태의 2차원 데이터의 크기를 색깔로 표시한다.

```python
from sklearn.datasets import load_digits

digits = load_digits()
X = digits.images[0]
X
```

```python
plt.title("mnist digits; 0")
plt.imshow(X, interpolation='nearest', cmap=plt.cm.bone_r)
plt.xticks([])
plt.yticks([])
plt.subplots_adjust(left=0.35, right=0.65, bottom=0.35, top=0.65)
plt.show()
```

* 인수
    * `cmap` : 데이터 수치를 색으로 바꾸는 칼라맵(color map)
        * 문자열 지정도 가능
        * 사용할 수 있는 칼라맵은 `plt.cm`의 속성으로 포함되어 있다.
        * https://matplotlib.org/tutorials/colors/colormaps.html

```python
dir(plt.cm)[:10]
```

```python
fig, axes = plt.subplots(1, 4, figsize=(12, 3), 
                         subplot_kw={'xticks': [], 'yticks': []})
axes[0].set_title("plt.cm.Blues")
axes[0].imshow(X, interpolation='nearest', cmap=plt.cm.Blues)

axes[1].set_title("plt.cm.Blues_r")
axes[1].imshow(X, interpolation='nearest', cmap=plt.cm.Blues_r)

axes[2].set_title("plt.BrBG")
axes[2].imshow(X, interpolation='nearest', cmap='BrBG')

axes[3].set_title("plt.BrBG_r")
axes[3].imshow(X, interpolation='nearest', cmap='BrBG_r')

plt.show()
```

* 인수
    * `interpolation` : 보간법

```python
methods = [None, 'none', 'nearest', 'bilinear', 'bicubic', 'spline16', 'spline36',
          'hanning', 'hamming', 'hermite', 'kaiser', 'quadric', 'catrom', 'gaussian',
          'bessel', 'mitchell', 'sinc', 'lanczos']
fig, axes = plt.subplots(3, 6, figsize=(12, 6),
                        subplot_kw={'xticks': [], 'yticks': []})

for ax, interp_method in zip(axes.flat, methods):
    ax.imshow(X, cmap=plt.cm.bone_r, interpolation=interp_method)
    ax.set_title(interp_method)
    
plt.show()
```

# 7. 컨투어 플롯
* 컨투어 플롯을 만들 때 입력 변수 x, y는 그대로 사용할 수 없고 `meshgrid` 명령으로 그리프 포인트 행렬을 만들어야 한다.

##### contour
등고선(contour)만 표시한다.
* 참고
    * http://matplotlib.org/api/pyplot_api.html#matplotlib.pyplot.contour

##### contourf
등고선을 색깔로 표시한다.
* 참고
    * http://matplotlib.org/api/pyplot_api.html#matplotlib.pyplot.contourf

```python
def f(x, y):
    return (1 - x / 2 + x ** 5 + y ** 3) * np.exp(-x ** 2 - y ** 2)

n = 256
x = np.linspace(-3, 3, n)
y = np.linspace(-3, 3, n)
XX, YY = np.meshgrid(x, y)
ZZ = f(XX, YY)
ZZ
```

```python
plt.title("Contour plots")
plt.contourf(XX, YY, ZZ, alpha=.75, cmap='jet')
plt.contour(XX, YY, ZZ, colors='black')
plt.show()
```

# 8. 3D 서피스 플롯
`Axes3D`라는 3차원 전용 axes를 생성하여 입체적으로 표시한다.

##### plot_wireframe
3D 서피스 플롯을 그린다.
* 참고
    * http://matplotlib.org/mpl_toolkits/mplot3d/api.html#mpl_toolkits.mplot3d.axes3d.Axes3D.plot_wireframe

##### plot_surface 
3D 서피스 플롯을 그린다.
* 참고
    * http://matplotlib.org/mpl_toolkits/mplot3d/api.html#mpl_toolkits.mplot3d.axes3d.Axes3D.plot_surface

```python
from mpl_toolkits.mplot3d import Axes3D

X = np.arange(-4, 4, 0.25)
Y = np.arange(-4, 4, 0.25)
XX, YY = np.meshgrid(X, Y)
RR = np.sqrt(XX ** 2 + YY ** 2)
ZZ = np.sin(RR)
```

```python
fig = plt.figure()
ax = Axes3D(fig)
ax.set_title("3D Surface Plot")
ax.plot_surface(XX, YY, ZZ, rstride=1, cstride=1, cmap='hot')
plt.show()
```

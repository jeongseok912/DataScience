# 1. 표본공간과 확률표본
* 확률표본(probabilistic sample, random sample) 또는 표본(sample) : 풀고자 하는 문제에서 발생(realize)할 수 있는 하나의 현상, 혹은 선택(sampled)될 수 있는 하나의 경우
* 표본공간(sample space) : 가능한 모든 표본의 집합. Ω(오메가)로 표기
    * 표본 공간을 정의하는 것 : 우리가 고려하는 범위에서 어떤 표본(경우, 현상)이 가능하고 어떤 표본이 불가능한지 정의하는 작업
    * 표본공간은 다음처럼 표기할 수 있다.
        * 동전 던지기 표본공간 - 앞면(Head), 뒷면(Tail)
        $$ Ω_1 = \{H, T\} $$
        * 플레잉카드 표본공간 - 다이아몬드, 하트, 스페이드, 클럽
        $$ Ω_2 = \{◆, ♥, ♠, ♣\} $$

# 3. 확률(probability)
사건(부분집합)을 입력하면 숫자(확률값)가 출력되는 함수
* 확률이라는 함수의 입력값의 집합 즉, 정의역(domain)은 표본공간의 모든 사건(부분집합)의 집합이다.(엄격한 정의의 시그마 대수(sigma algebra)는 고려하지 않는다.)
    $$ 확률 $$
$$ 사건 (부분집합) →→→→→→→→→ 숫자 $$
* 확률 : 모든 각각의 사건(부분집합)에 어떤 숫자를 할당(allocate)하는 함수
* 확률 P는 함수, P(A)는 A라는 사건(부분집합)에 할당된 숫자

#### 콜모고로프의 공리(Kolmogorov's axioms) = P(A)의 규칙
----
1. 모든 사건에 대해 확률은 실수이고, 0 또는 양수이다.
$$ P(A) ≥ 0 $$
2. 표본공간(전체집합)이라는 사건(부분집합)에 대한 확률은 1
$$ P(Ω) = 1 $$
3. 공통 원소가 없는 두 사건의 합집합의 확률은 각각의 사건의 확률의 합
$$ A ∩ B = /0 → P(A ∪ B) = P(A) + P(B) $$
----

* 표본공간이 Ω = {H, T}일 때,  
    * 사건(부분집합) {H}에 대한 확률값 P({H}) : "H라는 표본이 선택될 확률",  
    * 사건(부분집합) {H, T}에 대한 확률값 P({H, T}) : "H 또는 T라는 표본이 선택될 확률"

#### 확률이라는 함수를 정의한다는 것
"무엇이 선택될 확률이 얼마인가"라는 질문에 대한 답을 모든 경우(사건, 부분집합)에 대해 미리 준비해 놓은 것 또는 할당해 놓은 것

## 3.1. 확률은 표본이 아닌 사건을 입력으로 가지는 함수
* 확률은 "표본 하나 하나에 대해 정의되어 있는 숫자"가 아니다.
$$ 확률 $$
$$ 표본 →→→→→→→→→→→→ 숫자 (틀림!) $$
* 확률은 표본을 입력받아 숫자(확률값)를 출력하는 함수가 아니다.
$$ P(1) = 1/6 (틀림!) $$
* 확률은 표본이 아닌 사건(부분집합)에 대해 정의한다.
$$ P(\{1\}) = 1/6 $$

## 3.2. 확률 구현
사건을 key로 숫자를 value로 가지는 딕셔너리를 사용해 확률 함수를 구현할 수 있다.
* 이 때 key는 가능한 모든 부분집합이 되어야 한다.

```python
A = frozenset([]) 
B = frozenset(['H'])
C = frozenset(['T'])
D = frozenset(['H', 'T'])

P = {A: 0, B: 0.4, C: 0.6, D: 1}
P
```

* 앞면이 나오는 경우에 대해 할당된 확률이 0.5가 아닌 0.4  
→ 콜모고로프의 공리만 지킨다면 각각의 사건(부분집합)에 대한 확률값은 어떤 값을 할당해도 무관

#### 콜모고로프의 공리 만족 여부 확인
1. 모든 확률은 0 이상이다.
$$ P(/0) = 0 ≥ 0 $$
$$ P(\{H\} = 0.4 ≥ 0 $$
$$ P(\{T\}) = 0.6 ≥ 0 $$
$$ P(\{H, T\}) = 1 ≥ 0 $$
**⇒ 만족**
2. 전체집합에 대한 확률은 1이다.
$$ P(Ω) = P(\{H, T\}) = 1 $$
**⇒ 만족**
3. 교집합이 공집합인 사건(부분집합)의 합집합의 사건의 확률은 각 사건(부분집합)의 확률의 합이다.
$$ P(\{H\}) = 0.4 = P(\{H\} ∪ /0) = 0.4 + 0 $$
$$ P(\{T\}) = 0.6 = P(\{T\} ∪ /0) = 0.6 + 0 $$
$$ P(\{H, T\}) = 1 = P(\{H, T\} ∪ /0) = 1 + 0 $$
$$ P(\{H, T\}) = 1 = P(\{H\} ∪ \{T\}) = 0.4 + 0.6 $$
교집합이 공집합인 어떤 사건을 선택해도 합집합의 확률은 각 사건의 확률의 합과 같다.  
**⇒ 만족**

∴ 3가지 공리 모두 만족시키는 할당법이므로 위 함수는 확률이라고 부를 수 있다.

#### 골동품 가게 가격에 비유한 확률
* 가게에서 파는 하나 하나의 골동품은 표본이다. 모든 골동품은 서로 다르기 때문에 똑같은 물건(표본)은 없다.
* 가게에서 파는 모든 골동품의 집합은 표본공간(전체집합)이다.
* 사건(부분집합)은 골동품을 넣은 상자를 말한다. 
    * 상자 안의 골동품 갯수에는 제한이 없다. 골동품이 없는 빈 포장(공집합)도 가능하다.
* 확률은 상자에 붙인 가격표 숫자를 말한다. 가격은 마음대로 붙여도 되지만 다음 규칙을 지켜야 한다.
    1. 음수인 가격은 없다. 공짜(0)나 양수이어야 한다.
    2. 가게 안의 모든 골동품을 하나의 상자에 포장하면 그 상자의 가격은 1이다.
    3. 공통적으로 포함된 골동품이 없는 두 개의 상자의 가격은 그 두 개의 포장에 들어간 골동품을 합쳐서 하나의 상자로 만들었을 때의 가격과 같아야 한다. 즉 상자를 나누거나 합쳤다고 가격이 달라져서는 안된다.

## 3.3. 주사위 한 면의 확률은 무조건 1/6?
* 주사위의 어떤 한 면이 나올 확률이 1/6이라는 것은 주사위 확률의 하나의 예에 지나지 않으며 다른 확률도 가능하다. (
    * 아주 정밀하게 만들지 않으면 어떤 면이 미세하게나마 더 잘나오게 되는 경우가 발생할 수 있다.
* 극단적으로 도박에서 사기 치기 위해 특수한 주사기를 만들었다.  
이 주사위는 절대로 6이 나오지 않으며, 10번을 던지면 5번은 반드시 1이 나오게 만들었다.   
이 때 확률은 다음과 같이 할당되어 있다고 할 수 있다.
$$ P(\{1\}) = 0.5 $$
$$ P(\{6\}) = 0 $$
* 이 두 사건을 포함한 다른 모든 사건의 확률이 콜모고로프 정의를 만족한다면 이러한 확률값도 확률의 수학적 정의에 맞는다고 할 수 있다.

#### 공평한 주사위
* 1부터 4까지의 숫자만 나오는 4면체 주사위가 공평한 주사위라면 할당된 확률은?
$$ P(/0) = 0 $$
$$ P(\{1\}), P(\{2\}), P(\{3\}), P(\{4\}) = 1/4 $$
$$ P(\{1,2\}), P(\{1,3\}), P(\{1,4\}), P(\{2,3\}), P(\{2,4\}), P(\{3,4\}), = 1/2 $$
$$ P(\{1, 2, 3\}), P(\{1, 2, 4\}), P(\{1, 3, 4\}), P(\{2, 3, 4\}) = 3/4 $$
$$ P(\{1, 2, 3, 4\}) = 1 $$

#### 불공평한 주사위
* 절대로 1이 나오지 않도록 만들어진 4면체 주사위일 때 할당될 수 있는 확률은?
$$ P(/0) = 0 $$
$$ P(\{1\}) = 0, P(\{2\}) = 1/2, P(\{3\}) = 1/4, P(\{4\}) = 1/4 $$
$$ P(\{1, 2\}) = 1/2, P(\{1, 3\}) = 1/4, P(\{1, 4\}) = 1/4 $$
$$ P(\{2, 3\}) = 3/4, P(\{2, 4\}) = 3/4 $$
$$ P(\{3, 4\}) = 1/2 $$
$$ P(\{1, 2, 3\}) = 3/4, P(\{1, 2, 4\}) = 3/4, P(\{1, 3, 4\}) = 3/4, P(\{2, 3, 4\}) = 1 $$
$$ P(\{1, 2, 3, 4\}) = 1 $$

* 이 할당법은 콜모고로프 공리를 모두 만족한다.
1. 각 확률은 0보다 같거나 크다.
2. 전체집합에 할당된 확률은 1이다.
3. 교집합이 공집합인 모든 두 부분집합의 확률은 각 부분집합의 확률의 합이다.
$$ P(\{1, 2\}) = 1/2 = P(\{1\}) + P(\{2\}) = 0 + 1/2 $$
$$ P(\{1, 3\}) = 1/4 = P(\{1\}) + P(\{3\}) = 0 + 1/4 $$
$$ P(\{1, 4\}) = 1/4 = P(\{1\}) + P(\{4\}) = 0 + 1/4 $$
$$ P(\{2, 3\}) = 3/4 = P(\{2\}) + P(\{3\}) = 1/2 + 1/4 $$
$$ P(\{2, 4\}) = 3/4 = P(\{2\}) + P(\{4\}) = 1/2 + 1/4 $$
$$ P(\{3, 4\}) = 1/2 = P(\{3\}) + P(\{4\}) = 1/4 + 1/4 $$
$$ P(\{1, 2, 3\}) = 3/4 = P(\{1, 2\}) + P(\{3\}) = 1/2 + 1/4 $$
$$ P(\{1, 2, 3\}) = 3/4 = P(\{1\}) + P(\{2, 3\}) = 0 + 3/4 $$
$$ ... $$

두 조건 
$$ P(\{1\}) = 0.5 $$
$$ P(\{6\}) = 0 $$
을 만족하도록 주사위의 확률을 모든 사건(부분집합)에 대해 할당하고 이를 파이썬으로 구현한다.

----

### 3.3.1. 주사위의 한 면이 나올 확률이 무조건 1/6이라고 생각해 왔던 이유
* 확률의 정의와는 관계없는 "주사위가 공정한 주사위이다. 혹은 공정하지 않다고 생각할 수 있다는 증거가 아직 없다." 라는 사실을 가정하고 있기 때문

#### 공정한 주사위의 확률값
* 하나의 면이 나올 확률값은 같고, 이를 x라 할 때
$$ P(\{1\}) = P(\{2\}) = P(\{3\}) = P(\{4\}) = P(\{5\}) = P(\{6\}) = x $$
이 사건들은 서로 공통원소가 없고, 합집합은 전체 집합이므로  
확률의 성질을 이용하여 구하면 각 확률은 1/6이다.
$$ P(\{1\}) + P(\{2\}) + P(\{3\}) + P(\{4\}) + P(\{5\}) + P(\{6\}) = 6x = P(\{1, 2, 3, 4, 5\}) = P(Ω) = 1 $$
$$ x = 1/6 $$
* 이 방법은 확률 값을 만드는 방법의 하나일 뿐
    * 현실에서 꼭 이대로 확률 값 즉, 숫자를 배당할 이유는 없다.
    
#### 표본의 갯수가 유한하고 각 사건에 대해 원소의 갯수 이외의 아무런 정보가 없을 때의 확률
* 위 논리에 따르면 다음과 같다고 보는 것이 타당
$$ P(A) = card(A) / card(Ω) $$
* 그러나 다른 곳에서 얻은 정보(자연법칙 또는 도메인 지식 등)나 표본에 대한 데이터가 존재하는 경우 보다 믿을 수 있는 확률값을 계산할 수 있다.

#### 확률 사건의 예
20명의 사람 집합이 있다.
* 이 모임은 여자 + 남자로 구성
* 성별에 관계없이 머리가 짧은 사람과 긴 사람의 두 그룹으로 나눌 수 있다고 가정
* 이 경우 한 명의 사람은 표본, 전체 20명의 집합은 표본공간, 표본공간으로부터 몇 명의 사람을 선택하면 이 부분집합은 사건

#### 몇 가지 사건
* 전체 사람들 중 남자들의 모임은 부분집합이므로 사건A라고 하다.
* 전체 사람들 중 머리가 긴 사람들의 모임도 부분집합이므로 사건B라고 한다.
* 생일이 i월인 사람들의 집합도 부분집합이므로 사건이다. 1월인 사람의 집합은 C_1, 2월인 사람의 집합은 C_2, ... C_12까지의 집합을 만들 수 있다.
* 원소의 갯수가 0인 경우, 즉 원소가 없는 부분집합도 사건이다. 생일이 6월인 사람이 없으면 C_6의 원소 갯수는 0이다.

#### 확률 재정의
확률을 사건에 속하는 표본의 수로 정의해본다.
* 표본 집합에 20명의 사람(원소)이 있고, 어떤 사건에는 10명의 사람이 있다면 이 사건의 확률은 10/20 = 1/2
* 이 때는 "확률" = "사람의 수"라고 생각
* 이 방법은 확률을 정의하는 방법 중 하나 (반드시 확률이란 것이 이런 식으로 정해져야 하는 것은 아니다.)

# 확률의 성질
아래 성질은 많이 사용될 몇 가지 성질이다.  
이 성질들은 확률의 정의 자체가 아니라 정의로부터 유도된 것들

# 성질 1. 공집합의 확률
공집합인 사건의 확률은 0이다.
#### 증명
확률의 정의로부터 사건 A와 사건B가 공통원소가 없다면
$$ P(A ∪ B) = P(A) + P(B) $$
B = /0인 경우 A와 B의 공통원소는 없으며 A ∪ /0 = A라는 사실을 이용하면
$$ P(A ∪ /0) = P(A) = P(A) + P(/0) $$
$$ ∴ P(/0) = 0 $$

# 성질 2. 여집합의 확률
어떤 사건의 여집합인 사건의 확률은 (1 - 원래 사건의 확률)과 같다.
$$ P(A^C) = 1 - P(A) $$

#### 증명
확률의 정의로부터 사건 A와 사건B가 공통원소가 없다면
$$ P(A ∪ B) = P(A) + P(B) $$
B = A^C인 경우 A와 B의 공통원소는 없다.
$$ P(A ∪ A^C) = P(Ω) = 1 = P(A) + P(A^C) $$
$$ ∴ P(A^C) = 1 - P(A) $$
위 성질과 P(A^C)가 0보다 크거나 같아야 한다는 콜모고로프 공리 1을 결합하면 확률값은 1보다 작거나 같아야 한다.
$$ P(A) ≤ 1 $$

# 성질 3. 포함-배제 원리
* 포함-배제 원리(Inclusion-exclusion priciple) 혹은 덧셈 규칙(sum rule, addition law)   
두 사건의 합집합의 확률은 각 사건의 확률의 합에서 두 사건의 교집합의 확률을 뺀 것과 같다.
$$ P(A ∪ B) = P(A) + P(B) - P(A ∩ B) $$

#### 증명
$$ P(A ∪ B) = P(A ∪ (B ∩ A^C)) $$
$$ = P(A) + P(B ∩ A^C) $$
$$ = P(A) + P(B ∩ A^C) + P(A ∩ B) - P(A ∩ B) $$
$$ = P(A) + P((A^C ∩ B) ∪ (A ∩ B)) - P(A ∩ B) $$
$$ = P(A) + P(B) - P(A ∩ B) $$

# 성질 4. 전체 확률의 법칙
* 복수의 사건 C_i가 다음을 만족하는 사건들인 경우,  
서로 교집합이 없고
$$ C_i ∩ C_j = /0 (i ≠ j) $$
모두 합쳤을 때 (합집합이) 전체 표본공간이면
$$ C_1 ∪ C_2 ∪ ... = Ω $$
* 전체 확률의 법칙(law of total probability)  
모든 사건 A에 대해 사건 A의 확률은 사건 A와 사건 C_i가 동시에 발생할 사건들의 확률의 합과 같다.
$$ P(A) = ∑_iP(A ∩ C_i) $$
* 위 예에서 사건 P(A ∩ C_1)은 생일이 1월인 남자의 집합이고, 전체 확률의 법칙은 다음과 같이 이해할 수 있다.
    * 생일이 1월인 남자라는 사건의 확률과 2월인 남자라는 사건의 확률, 이외에도 각 월을 생일인 남자라는 사건들의 확률을 모두 합치면 남자라는 사건의 확률이 된다.
    
#### 증명
$$ A = A ∩ Ω $$
$$ = A ∩ (C_1 ∪ C_2 ∪ ...) $$
$$ = (A ∩ C_1) ∪ (A ∩ C_2) ∪ ... $$
A ∩ C_i는 모두 서로 공통 원소가 없다.   
따라서 확률의 정의에 따라 다음 등식이 성립
$$ P(A) = P(A ∩ C_1) + P(A ∩ C_2) + ... = ∑_iP(A ∩ C_i) $$
교집합의 확률은 쉼표로 표시하기도 한다.
$$ P(A ∩ B) = P(A, B) $$
따라서 전체확률의 법칙은
$$ P(A) = ∑_iP(A, C_i) $$

# 연습 문제
앞 연습 문제에서 정한 확률 값을 이용하여 확률의 성질2, 3, 4번이 성립하는지 확인해 본다.
* 성질 2. 여집합의 확률
    * 1 - P(A) = 1 - 0.6 = 0.4
    → 여자는 8명
    * 1 - P(B) = 1 - 0.3 = 0.7
    → 머리가 짧은 사람은 14명
    * 1 - P(C) = 1- 1 = 0
    → 모두 생일 정보가 있다.
* 성질 3. 포함-배제의 원리
    * P(A ∪ B) = P(A) + P(B) - P(A ∩ B)  
              * 머리가 길고 남자인 사람 2명이라고 가정.   
              * P(A ∩ B) = 2/20 = 0.1
      * P(A) + P(B) - P(A ∩ B) = 0.6 + 0.3 - 0.1 = 0.8
      * P(A U B) = 18/20 = 0.9

# 정리
* 공집합의 확률
$$ P(/0) = 0 $$
* 여집합의 확률
$$ P(A^C) = 1 - P(A) $$
$$ 0 ≤ P(A) ≤ 1 $$
* 포함-배제 원리
$$ P(A ∪ B) = P(A) + P(B) - P(A ∩ B) $$
* 전체 확률의 법칙
$$ P(A) = ∑_iP(A ∩ C_i) $$

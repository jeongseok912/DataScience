# 1. 빈도주의 관점에서 확률의 의미
* 빈도주의(frequentist)에서는 반복적으로 선택된 표본이 사건(부분 집합) A의 원소가 될 경향(propensity)을 그 사건의 확률이라고 한다.  
ex)  
동전을 던져 "앞면이 나오는 사건"의 확률값이 0.5라는 것 : 빈도주의 관점에서는 동전을 반복하여 던졌을 경우 동전을 던진 전체 횟수에 확률값을 곱한 숫자만큼 해당 사건이 발생한다고 본다. 10,000번 던지면 10,000 * 0.5 = 5,000번 앞면이 나오는 경향을 가진다는 의미

# 2. 베이지안 관점에서 확률의 의미
* 베이지안(bayesian) 관점에서 확률은 "선택된 표본이 특정한 사건(부분 집합)에 속한다는 가설(hypothesis), 명제(proposition) 혹은 주장(assertion)의 신뢰도(degree of belief)"라고도 볼수 있다. 여기에 반복이라는 개념은 사용되지 않는다.  
ex1)  
"새는 날수 있다" 라는 명제가 있을 때 이 명제는 참인가? 닭, 타조, 펭귄 등의 새나 어린 새, 날개를 다친 새는 날 수 없으므로 항상 참인 명제는 아니다.  
→ 베이지안 관점에서 이 명제를 다시 서술하면 "새가 날 수 있는 가능성은 95%다."  
ex2)  
"동전 앞면이 나오는 사건"의 확률값이 0.5다. → 베이지안 관점에서는 "앞면이 나왔다"는 주장의 신뢰도가 0.5다.
* 4지 선다형 객관식 문제를 풀 때 우리는 베이지안 확률을 사용한다. 보기를 보고 각 보기에 대한 확률을 할당하기 때문
    > '1번은 절대로 답이 될 수 없다. 그러니까 1번이 정답일 확를은 0이다.  
    2번과 3번이 그럴 듯한데. 4번은 가능성이 2번이나 3번의 가능성의 반도 안되어 보이고.  
    그러니까 2번과 3번이 정답일 확률은 각각 0.4이고 4번이 정답일 확률은 0.2 이군.'
* 그러면 {1, 2, 3, 4}라는 표본집합이 있을 때 할당한 확률은
$$ P(\{1\} = 0 $$
$$ P(\{2\} = 0.4 $$
$$ P(\{3\} = 0.4 $$
$$ P(\{4\} = 0.2 $$
    * 여기서 확률의 정의는 반복과 빈도와는 전혀 관계없다. 확률 P({1})은 "정답은 1이다"라는 주장에 대한 신뢰도일 뿐이다.

# 3. 사건의 발생
* 베이지안 관점에서 사건(부분 집합) : "원하는 답(표본)이 포함되어 있을 가능성이 있는 후보의 집합"
    * 우리가 어떤 사건을 제시하면 그 자체로 "이 사건에 속한 원소 중에 원하는 답(표본)이 있다"는 명제 혹은 주장을 제시한 것이라 할 수 있다.
* 베이지안 확률론에서 사건이 일어났다(occur) 혹은 발생했다는 말 : 그 사건의 원소 중에 정말로 선택된 표본이 있다는 사실을 알게 되었다는 것을 말한다.
    * 다른 말로 해당 사건이 말하고 있는 주장이 진실임을 알게 되었다는 뜻  
    (여태 모르던 추가적인 정보가 들어왔음을 뜻함)  
ex)  
불투명 컵 안에 주사위를 굴렸다고 가정했을 때 "눈금이 짝수가 나오는 사건이 발생했다"라는 말은 컵을 들어서 눈금을 본 사람이 "주사위의 눈금이 짝수다"라는 사실인 정보를 준 것을 의미한다.

# 4. 빈도주의적 관점과 베이지안 관점의 비교
* 확률의 빈도주의적 관점과 베이지안 관점은 양립할 수 없는 관계가 아니다.  
ex)  
"검진 결과로 보아 암에 걸렸을 확률이 90%이다"라고 진단했다 가정했을 때 
    * 의사의 관점 : "이런 검진 결과를 가진 환자를 정밀 검사로 확인하는 일을 100번 반복하면 그 중의 약 90명은 암에 걸려있다."라는 빈도주의적일 수 있다.
    * 환자의 관점 : 다른 환자가 암인가 아닌가는 의미가 없기 때문에 "자신이 암에 걸렸다는 의사의 주장이 사실일 가능성은 90%이다."라는 베이지안 관점으로 바라보게 된다.

# 5. 확률분포의 정보성
* 베이지안 관점에서 보면 알아낸 확률분포는 표본이 어떤 상태에 있는지를 알려주는 정보이다.
* 4지 선다형 객관식 문제의 확률이 한 사건에 몰려있다고 하면
$$ P(\{1\}) = 0 $$
$$ P(\{2\}) = 0 $$
$$ P(\{3\}) = 1 $$
$$ P(\{4\}) = 0 $$
"답이 3이다."라는 정보를 확보한 것과 마찬가지  
따라서 확률분포를 알아내면 문제의 답을 알고 있는지 모르는지를 보여줌과 동시에 만약 알고 있다면 답이 무엇이다라는 정보까지 주게 된다.
* 반대로 확률이 고르게 분포되어 있다면
$$ P(\{1\}) = 0.25 $$
$$ P(\{2\}) = 0.25 $$
$$ P(\{3\}) = 0.25 $$
$$ P(\{4\}) = 0.25 $$
이 문제에 대해 아무것도 모른다. 즉 아무런 정보다 없다고 고백하는 것과 마찬가지이다.  
이러한 확률분포의 정보성은 엔트로피(entropy)라는 측도로 측정하게 된다.

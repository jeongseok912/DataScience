# 1. 확률분포의 기댓값
* 기댓값(expectation) : 확률변수의 확률밀도함수 $p(x)$를 알면 확률변수의 표본평균의 이론적인 값을 구할 수 있다. 즉, 확률변수의 이론적인 표본평균
    * 경우에 따라 단순히 평균(mean)이라고도 한다.
    * 확률변수의 기댓값을 구하는 연산자는 $E[·]$ 로 표기
    * 기댓값은 $\mu_X$로 표기 (확률변수 혼동 염려 없을 땐 그냥 $\mu$로 써도 무방)

## 1.1. 이산확률변수의 기댓값
표본공간의 원소 $x_i$의 가중평균  
* 이 때 가중치는 $x_i$가 나올 수 있는 확률.  
즉, 확률질량함수 $P(x_i)$이다.
$$ \mu_X = E[X] = \sum_{x_i ∈ Ω}x_iP(x_i) $$
<img src="./dsschool_img/discrete_E[X].png">

ex)  
공정한 주사위에서 나올 수 잇는 숫자를 대표하는 확률변수 $X$는 나올 수 있는 값이 1 ~ 6이므로,
$$ \mu_X = 1·\frac{1}{6} + 2·\frac{1}{6} + 3·\frac{1}{6} + 4·\frac{1}{6} + 5·\frac{1}{6} + 6·\frac{1}{6} = 3.5 $$
기댓값은 3.5이다.

## 1.2. 연속확률변수의 기댓값
확률밀도함수 $p(x)$를 가중치로 $x$를 적분한 값
$$ \mu_X = E[X] = ∫_{-\infty}^\infty xp(x)dx $$
<img src="./dsschool_img/continuous_E[X].png">

ex)  
회전하는 원반을 이용해서 복권 번호를 결정하는 문제에서 확률밀도함수 $p(x)$와 $xp(x)$의 모양은 다음과 같다.  
기댓값은 이 함수 $xp(x)$를 적분하여 구한 면적

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-100, 500, 1000)
p = np.zeros_like(x)

p[(0 < x) & (x <=360)] = 1/360
xp = x * p

plt.subplot(121)
plt.plot(x, p)
plt.xticks([0, 180, 360])
plt.title("$p(x)$")
plt.xlabel("$x$ (deg.)")

plt.subplot(122)
plt.plot(x, xp)
plt.xticks([0, 180, 360])
plt.title("$xp(x)$")
plt.xlabel("$x$ (deg.)")

plt.show()
```

```python
# 0 ~ 180˚ 사이에 화살이 2배 더 잘박히도록 조작된 원반의 확률밀도함수와 기댓값
x = np.linspace(-100, 500, 1000)
p = np.zeros_like(x)

p[(0 < x) & (x <= 180)] = 2 / (3 * 360)
p[(180 < x) & (x <= 360)] = 1 / (3 * 360)
xp = x * p

plt.subplot(121)
plt.plot(x, p)
plt.xticks([0, 180, 360])
plt.title("$p(x)$")
plt.xlabel("$x$ (deg.)")

plt.subplot(122)
plt.plot(x, xp)
plt.xticks([0, 180, 360])
plt.title("$xp(x)$")
plt.xlabel("$x$ (deg.)")

plt.show()
```

# 연습 문제
확률변수 $X$는 주사위를 던져 나오는 수를 나타내는 확률변수이다. 그리고 $Y$는 주사위를 던져나오는 수에 2배를 한 수를 나타내는 확률변수이다. $X, Y$의 확률질량함수의 그래프를 각각 그려라.

```python
x = np.arange(1, 7)
px = np.array([1/6, 1/6, 1/6, 1/6, 1/6, 1/6])
py = np.array([2/6, 2/6, 2/6, 2/6, 2/6, 2/6])

plt.subplot(121)
plt.plot(x, px)
plt.xticks([1, 2, 3, 4, 5, 6])
plt.grid()

plt.subplot(122)
plt.plot(x, py)
plt.xticks([1, 2, 3, 4, 5, 6])
plt.grid()

plt.tight_layout()
plt.show()
```

다만 확률변수 $X$에서 표본을 $N$번 뽑아서 연산하는 경우에는 같은 확률변수이지만 표본값이 다르기 때문에 원래 확률변수의 복사본 $X_1, X_2, ..., X_N$을 만든 다음 이 확률변수의 표본값을 더한 형태로 변환식을 써야 한다.
$$ Y = X_1 + X_2 + ... + X_N $$
* 이렇게 쓰는 이유  
$X_1$과 $X_2$가 같은 확률변수이지만 표본값이 다르기 때문
* 만약 다음같이 쓰게 된다면
$$ Y = X + X + \dots + X $$
이 식은 전현 다른 확률변수를 가리킨다.
$$ Y = N·X $$

# 5. 기댓값의 성질
* 변환된 확률변수의 기댓값을 계산할 때는 기댓값의 성질을 이용한다.
    * 확률변수가 아닌 상수 $c$에 대해 
    $$ E[c] = c $$
    * 선형성
    $$ E[cX] = cE[X] $$
    $$ E[X + Y] = E[X] + E[Y] $$
    $$ E[c_1X + c_2Y] = c_1E[X] + c_2E[Y] $$

# 6. 표본평균의 확률분포
* 확률변수로부터 $N$개의 표본을 만들어 이 표본집합의 표본평균을 구하면 이 표본평균 값도 확률변수이다.
* 표본평균의 확률변수는 원래의 확률변수 이름에 bar를 추가하여 $\bar X$와 같이 표기  
* 확률변수 $X$에서 나온 표본으로 만들어진 표본평균의 확률변수 $\bar X$
$$ \bar X = \frac{1}{N}\sum_{i=1}^NX_i $$
    * $X_i$는 $i$번째로 실현된 표본값을 생성하는 확률변수
    * 이 확률변수 $X_i$는 원래의 확률변수 $X$의 복사본

# 7. 기댓값과 표본평균의 관계
* 표본평균도 확률변수이므로 기댓값이 존재한다.
* 표본평균의 기댓값은 원래의 확률변수의 기댓값과 같다
$$ E[\bar X] = E[X] $$
> **증명**
$$ E[\bar X] = E[\frac{1}{N}\sum_{i=1}^NX_i] $$
$$ = \frac{1}{N}\sum_{i=1}^NE[X_i] \text{ (선형성)} $$
$$ = \frac{1}{N}\sum_{i=1}^NE[X] \text{ (}X_i\text{는 X의 복사본이므로 기댓값이 같다.)} $$
$$ = \frac{1}{N}NE[X] \text{ (동일한 값의 합)} $$
$$ = E[X] $$

#### 의미
표본평균은 확률변수의 기댓값 근처의 값이 된다.

# 8. 중앙값
* 확률변수의 중앙값은 중앙값보다 큰 값이 나올 확률과 작은 값이 나올 확률이 0.5로 같은 값을 뜻함
* 따라서 누적확률분포 $F(x)$에서 중앙값을 계산할 수 잇다.
$$ 0.5 = F(median) $$
$$ median = F^{-1}(0.5) $$
